```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr
from mpl_toolkits.basemap import Basemap
```

```python
peta = Basemap(projection = 'cyl', llcrnrlon = 95, llcrnrlat = -11, urcrnrlon = 141, urcrnrlat = 7, resolution = 'i')
peta.drawcoastlines(1)
peta.drawcountries()
peta.bluemarble()

parallels = np.arange(-11, 7 + 0.25, 3)
peta.drawparallels(parallels, labels = [1,0,0,0], linewidth = 0.5)
meridians = np.arange(95, 141 + 0.25, 5)
peta.drawmeridians(meridians, labels = [0,0,0,1], linewidth = 0.5)

plt.show()
# plt.title('Peta Indonesia')
```
